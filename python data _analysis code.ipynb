###SocialPost class with encapsulation (private attributes for
likes/comments)

````python
class SocialPost:
    def __init__(self, post_id, category, author, likes, comments):
        self.post_id = post_id
        self.category = category
        self.author = author
        self.__likes = likes
        self.__comments = comments

    def get_private_accsess(self):
        return self.__likes + self.__comments

    def display(self):
        print(f"Post ID: {self.post_id}, Category: {self.category}")
````

Author subclass inheriting from SocialPost with additional author_rating
property


````python
class Author(SocialPost):
    def __init__(self, post_id, category, author, likes, comments, author_rating):
        super().__init__(post_id, category, author, likes, comments)
        self.author_rating = author_rating  # New property

    def display(self):
        print(f"Author: {self.author}, Rating: {self.author_rating}, Category: {self.category}")
````

demonstrate to polymorphism in display method

````python

post1 = SocialPost("1", "Tech", "tulasi", 120, 40)
post2 = Author("2", "Food", "rupa", 300, 55, 4.7)

for post in [post1, post2]:
    post.display()
````

Part D: Data Visualization

Line chart showing daily post frequency

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# a) Line chart: Daily post frequency
df['Date_Posted'] = pd.to_datetime(df['Date_Posted'])
daily_posts = df['Date_Posted'].value_counts().sort_index()
daily_posts.plot(kind='line', figsize=(10, 5), title='Daily Post Frequency')
plt.xlabel('Date')
plt.ylabel('Number of Posts')
plt.grid(True)
plt.show()
````

Stacked bar chart of average engagement by category

````python
df['Engagement'] = df['Likes'] + df['Shares'] + df['Comments']
engagement = df.groupby('Category')[['Likes', 'Shares', 'Comments']].mean()
engagement.plot(kind='bar', stacked=True, figsize=(8, 5), title='Avg Engagement by Category',color=['red','green','purple'])
plt.ylabel("Average Count")
plt.show()
````

Seaborn heatmap of correlation matrix

````python

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True,color='green')
plt.title('Correlation Heatmap')
plt.show()
````
